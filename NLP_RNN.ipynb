```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
tf.random.set_seed(42)
```

```python
h = [1, 0, 0, 0]
e = [0, 1, 0, 0]
l = [0, 0, 1, 0]
o = [0, 0, 0, 1]
```

```python
x_data = np.array([[h]], dtype=np.float32)
x_data
```

```python
x_data.shape # (sample의 갯수, token의 갯수,  token의 embedding의 수) => (1, 1, 4)
```

```python
# input dimension : 4
# output dimension : 2 -> hidden_size로 세팅 -> 몇개의 클래스로 분류하는지에 따름

hidden_size = 2
cell = layers.SimpleRNNCell(units = hidden_size)
rnn = layers.RNN(cell, return_sequences = True, return_state= True)
output, states = rnn(x_data)

print('x_data : {} \t shape : {}'.format(x_data, x_data.shape))
print('output : {} \t shape : {}'.format(output, output.shape))
print('states : {} \t shape : {}'.format(states, states.shape))
```

```python
# vocab
idx2char = ['토', '마', '를', '먹', '자']

x_data = [[0, 0, 1, 2, 4, 3]] # 토 토 마 를 자 먹
y_data = [[0, 1, 0, 2, 3, 4]] # 토 마 토 를 먹 자

# idx2char 갯수
num_classes = 5

# 단어의 갯수 (문장에서 단어를 추출할 때는 보통 num_classes = input_dim)
input_dim = 5

# data의 갯수
sequence_len = 6

learning_rate = 0.1

x_one_hot = tf.keras.utils.to_categorical(x_data, num_classes=num_classes)
y_one_hot = tf.keras.utils.to_categorical(y_data, num_classes=num_classes)

print(x_one_hot)
print(x_one_hot.shape)
print(y_one_hot)
print(y_one_hot.shape)
```

```python
from tensorflow.keras.models import Model

model = tf.keras.Sequential()
cell = layers.SimpleRNNCell(units=num_classes, input_shape=(sequence_len, input_dim))


model.add(layers.RNN(cell=cell, return_sequences=True, return_state=False, input_shape= (sequence_len, input_dim)))
model.add(layers.TimeDistributed(layers.Dense(units=num_classes, activation='softmax')))

model.compile(loss='categorical_crossentropy', 
              optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(x_one_hot, y_one_hot, epochs=10)
```

```python
pred = model.predict(x_one_hot)
pred
```

```python
for i, word in enumerate(pred):
    print([idx2char[c] for c in np.argmax(word, axis=1)])
```
